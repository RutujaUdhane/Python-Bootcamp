```python
import random

def display(board):
    print(f'''
{board[7]}|{board[8]}|{board[9]}                      7 | 8 | 9
---+---+---                     ---+---+---
{board[4]}|{board[5]}|{board[6]}         Positions->  4 | 5 | 6
---+---+---                     ---+---+---
{board[1]}|{board[2]}|{board[3]}                      1 | 2 | 3
    ''')
    
def valid_input_pos(turn):
    while True:
        pos = input("Enter position : ")
        if pos.isnumeric() and int(pos) in range(1,10):
            pos = int(pos)
            break
        else: print('Not a valid position. Make sure that your input lies in the range 1 to 9')
    return int(pos)    

def board_update(turn,board):
    print(turn+"chance")
    pos = valid_input_pos(turn)
    while True:
        if board[pos]=='   ':
            board[pos]= turn
            break
        else:
            print('Cannot Overwrite, please select an other location which is not already occupied.')
            pos= valid_input_pos(turn)
            
def check(board):
    check=0
    if board[1]==board[2]==board[3]!='   ' or board[4]==board[5]==board[6]!='   ' or board[7]==board[8]==board[9]!='   ':
        check=1
    elif board[1]==board[4]==board[7]!='   ' or board[2]==board[5]==board[8]!='   ' or board[3]==board[6]==board[9]!='   ':
        check =1
    elif board[1]==board[5]==board[9]!='   ' or board[3]==board[5]==board[7]!='   ':
        check = 1
    return check    

def userInput(board, symbol):
    sym1, sym2 = symbol[random.randint(0,1)]
    print(f'{sym1} is going first\n\n')
    display(board)
    for i in range(9):
        if i%2==0:
            turn = ' '+sym1+' '
            board_update(turn,board)
        else:
            turn = ' '+sym2+' '
            board_update(turn,board)
        display(board)
        if i>=4:
            if check(board):
                display(board)
                print(turn+"WON")   
                break
            if i == 8:
                print("Nobody WON, it's a TIE!")
    
def main():
    board = ["Just to adjust location :",'   ','   ','   ','   ','   ','   ','   ','   ','   ']
    symbol = [('X','O'),('O','X')]
    while True:
        marker= input("\nEnter your marker (X/O): ").upper()
        if marker=='X' or marker=='O':
            userInput(board,symbol)
            if input("If you want to play again press 'Y' (press any other key to exit ): ").upper()=='Y':
                board = ["Just to adjust location :",'   ','   ','   ','   ','   ','   ','   ','   ','   ']
                pass
            else:
                print("Thank you for playing Tic-Tac-Toe")
                break
        else:
            print("Oops, wrong marker! Please enter again and make sure that your marker is either 'X' or 'O'.\n")
main()    
   
```
